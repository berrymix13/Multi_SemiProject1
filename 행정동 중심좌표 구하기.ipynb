```python
json = json.load(open('./json/서울행정동_수정.json','r', encoding = 'utf-8'))
```

```python
features = pd.DataFrame(json['features'])
features = features[['properties', 'geometry']]
```

```python
features.info()
```

```python
allavg=[]
for f in range(len(features)):
    lonlatlst=features.loc[f, 'geometry']['coordinates'][0][0]
    lonsum=0
    latsum=0
    for i in range(len(lonlatlst)):
        lonsum+=lonlatlst[i][0]
        lonavg=lonsum/len(lonlatlst)
        latsum+=lonlatlst[i][1]
        latavg=latsum/len(lonlatlst)
        lonlatavg=[latavg,lonavg]
    allavg.append(lonlatavg)
allavg
```

```python
df = pd.DataFrame(allavg)
```

```python
donglst=[]
for f in range(len(features)):
    dong=features.loc[f, 'properties']['동']
    donglst.append(dong)
dongdf=pd.DataFrame(donglst)
```

```python
df2=pd.concat([dongdf,df],axis=1)
df2.columns = ["동", "Latitude","Longitude "]
df2
```
