```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def request_url(urldate, endpage=999):   # url 생성 함수
    user_key="464c447675666c75373156546a786e"
    url="http://openapi.seoul.go.kr:8088/" + user_key
    url=url + "/xml/disabledCalltaxi/1/"+ str(endpage) # 출력 데이터 개수
    url=url + "/" + str(urldate)    # 조회 날짜
    #print(url)
    taxi_html=requests.get(url)  # 생성된 url을 기준으로 요청

    if taxi_html.status_code != 200: # 데이터를 받아오지 못했다면 종료
        exit("데이터를 받지 못함")
    
    return taxi_html

def taxi(schdate):   # 요청데이터를 이용해 DataFrame으로 제공 함수
    taxi_html=request_url(schdate)
    #print(taxi_html.text)
    soup=bs(taxi_html.text, "html.parser")
    items=soup.find_all("item")

    sw=[]
    for i in items:
        car_num=i.find('no').get_text()
        car_type=i.find('cartype').get_text()
        receipt_time=i.find('receipttime').get_text()
        set_time=i.find('settime').get_text()
        ride_time=i.find('ridetime').get_text()
        start_pos1=i.find('startpos1').get_text()
        start_pos2=i.find('startpos2').get_text()
        end_pos1=i.find('endpos1').get_text()
        end_pos2=i.find('endpos2').get_text()

        sw.append([car_num,car_type, receipt_time,set_time,ride_time,start_pos1, start_pos2,end_pos1,end_pos2])
    swdf=pd.DataFrame(sw, columns=["차량고유번호", "차량타입", "예정일시", "배차일시", "승차일시","출발지구군","출발지상세","목적지구군","목적지상세" ]) 
    return swdf

```
