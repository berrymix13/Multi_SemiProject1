```python
def speed():
    day=input("날짜입력-예시:2022.01.01")
    time=input("시간입력-예시:1시")
    map = folium.Map(location = [37.5502, 126.982], zoom_start=11)
    sdf=speed_df[speed_df['일자']==day]
    sdf=sdf[['name',time]]
    fig = folium.Choropleth(geo_data=geo_str, 
                        data=sdf ,
                        columns=['name',time],
                        key_on='feature.properties.name',
                        fill_color='YlGn',
                        fill_opacity=0.7,
                        line_opacity=0.3,
                        legend_name='속도').add_to(map)
                        

    return map
```
